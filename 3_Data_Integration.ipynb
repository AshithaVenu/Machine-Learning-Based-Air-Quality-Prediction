```python
import pandas as pd

# === 1️⃣ Load your main dataset ===
main_path = "/mnt/c/Users/janat/Documents/data science_UCL/cleaned-data-datascience-louvain-pollution-2025/merged_all.csv"
main_df = pd.read_csv(main_path)

# Rename columns to match the AOD dataset
main_df = main_df.rename(columns={"GRID": "grid_id", "MONTH": "month"})

print("📋 Columns in main dataset after renaming:")
print(main_df.columns.tolist())
print("\nSample rows:")
print(main_df.head())

# === 2️⃣ Load the merged AOD CSV ===
aod_path = "/mnt/c/Users/janat/Desktop/vs/correct/AOD_merged_all_types.csv"
aod_df = pd.read_csv(aod_path)

print("\n📋 Columns in AOD dataset:")
print(aod_df.columns.tolist())
print("\nSample rows:")
print(aod_df.head())

# === 3️⃣ Merge based on grid_id and month ===
merge_keys = ["grid_id", "month"]

merged_df = main_df.merge(aod_df, on=merge_keys, how="left")

# === 4️⃣ Save to CSV ===
output_path = "/mnt/c/Users/janat/Desktop/vs/MERGE_ALL.csv"
merged_df.to_csv(output_path, index=False)

print(f"\n✅ Done! Saved merged file to: {output_path}")
print("New columns added:")
print([c for c in merged_df.columns if "AOD" in c])
print("\nFirst few rows of merged data:")
print(merged_df.head())
```
